## 3) Research Questions (RQs)

### RQ1
**(a) What is your research question?**  
When two concepts are **combined into a single AI-generated image** (concept fusion), does this systematically change **memorability** compared to a baseline? Specifically, do memorability gains/losses vary by a **22 design**: **semantic similarity** (High vs. Low)  **category relation** (Same vs. Different category)?

**(b) What data and data mining techniques are you employing?**  
I construct concept pairs and assign them into four groups (High/Low semantic  Same/Different category). For each pair, I generate **combined images** and create a **baseline control** (copy/non-fused) condition. I use **CLIP** embeddings to operationalize semantic similarity and to support filtering/quality checks. I estimate image memorability using **ResMem** (model-based memorability proxy). I compile a structured dataset containing memorability outcomes plus covariates (e.g., CLIP features/similarity scores, generation metadata, filtering indicators), then fit **regression models** with main effects and interaction terms aligned to the 22 design.

**(c) What are your findings and how will you report them (or what remains)?**  
The end-to-end pipeline (generation  CLIP  filtering  ResMem  analysis table) is functional. Remaining work is to finalize the control/baseline comparison at scale and complete robustness checks (e.g., alternative similarity definitions, filtering thresholds).  
Planned reporting (within IC2S2 constraints):
- **Figure 1**: Framework/pipeline diagram (end-to-end workflow)
- **Figure 2**: Memorability distributions by the 4 experimental groups (box/violin)
- **Figure 3**: Regression coefficients (forest plot) emphasizing interaction effects
- **Table 1 (optional)**: Key model summary / robustness checks (or move to GitHub if space is tight)

**(d) What is the impact of answering this research question?**  
**Theoretical contribution**: This project links memorability to **how cultural/semantic boundaries are recombined**concept fusion may increase distinctiveness or schema incongruity, making some combinations more salient and thus more memorable.  
**Methodological contribution**: I provide a reproducible **multimodal data-mining pipeline** that treats AI-generated images as controllable cultural objects, quantifies semantic structure via CLIP, and measures memorability via a scalable proxy (ResMem), enabling systematic tests of how semantic organization relates to memory outcomes.

## 4) Summary: How are these RQs related?
RQ1 provides a controlled, hypothesis-driven test of how **semantic similarity and category boundaries** shape memorability (and memorability change vs baseline).
